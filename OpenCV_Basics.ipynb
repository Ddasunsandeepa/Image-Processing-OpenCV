```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('flower.jpg')
```

```python
print(image.shape)
```

```python
print(image[100, 0])
```

```python
from google.colab.patches import cv2_imshow

image[100:120,100:180] = [0,255,255]
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(image[:5, :5, :])
```

```python

# Load image using OpenCV (this reads as BGR)
img = cv2.imread('flower.jpg')

plt.imshow(img)
plt.axis('off')   # remove axis
plt.show()
# Convert BGR → RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_rgb[100:120,100:180] = [0,255,255]

plt.imshow(img_rgb)
plt.axis('off')   # remove axis
plt.show()
```
