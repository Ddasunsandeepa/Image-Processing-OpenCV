```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('orange.jpg')  # sample image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,7))
plt.subplot(2,3,1); plt.imshow(img); plt.title("BGR Image"); plt.axis('off')
plt.subplot(2,3,2); plt.imshow(img_rgb); plt.title("Original (RGB)"); plt.axis('off')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(2,3,3); plt.imshow(gray, cmap='gray'); plt.title("Grayscale"); plt.axis('off')

edges = cv2.Canny(gray, 100, 200)
plt.subplot(2,3,4); plt.imshow(edges, cmap='gray'); plt.title("Canny Edges"); plt.axis('off')

plt.show()
```

```python
from skimage import data, filters
import matplotlib.pyplot as plt   # <-- missing line

img = data.coins()
edges = filters.sobel(img)
blurred = filters.gaussian(img, sigma=2)

fig, ax = plt.subplots(1, 3, figsize=(12,4))
ax[0].imshow(img, cmap='gray'); ax[0].set_title("Original")
ax[1].imshow(edges, cmap='gray'); ax[1].set_title("Sobel Edges")
ax[2].imshow(blurred, cmap='gray'); ax[2].set_title("Gaussian Blur")

for a in ax:
    a.axis('off')

plt.show()
```

```python
# --- Pillow Example ---

from PIL import Image, ImageEnhance, ImageFilter

im = Image.open('orange.jpg')
enhancer = ImageEnhance.Contrast(im)
im_enhanced = enhancer.enhance(1.5)
im_blur = im.filter(ImageFilter.BLUR)

fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].imshow(im); ax[0].set_title("Original")
ax[1].imshow(im_enhanced); ax[1].set_title("Enhanced Contrast")
ax[2].imshow(im_blur); ax[2].set_title("Blurred")
for a in ax: a.axis('off')
    
plt.show()
```

```python
# --- NumPy Pixel Manipulation ---

import numpy as np

arr = np.array(im)
gray = np.dot(arr[...,:3], [0.2989, 0.5870, 0.1140])
mask = gray > 128
highlight = arr.copy()
highlight[~mask] = 0

fig, ax = plt.subplots(1,2, figsize=(10,4))
ax[0].imshow(gray, cmap='gray'); ax[0].set_title("Manual Grayscale")
ax[1].imshow(highlight); ax[1].set_title("Masked (Gray>128)")
for a in ax: a.axis('off')
    
plt.show()

print(arr.shape)   # (512, 512, 3)
print(arr[0,0])    # [R, G, B] values of the top-left pixel
```

```python
# --- SciPy ndimage Example ---

from scipy import ndimage

img = data.coins()
blur = ndimage.gaussian_filter(img, sigma=4)
edges = ndimage.sobel(img)

fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].imshow(img, cmap='gray'); ax[0].set_title("Original")
ax[1].imshow(blur, cmap='gray'); ax[1].set_title("Gaussian Blur")
ax[2].imshow(edges, cmap='gray'); ax[2].set_title("Sobel Edge")
for a in ax: a.axis('off')
    
plt.show()
```
