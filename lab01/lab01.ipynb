```python
plt.hist(gray.ravel(), 256, [0,256], color='gray', alpha=0.7, label='Original')
plt.title("Histogram")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

# Apply histogram equalization
equalized = cv2.equalizeHist(gray)

# Plot histograms
plt.hist(gray.ravel(), bins=256, range=(0,256), color='gray', alpha=0.6, label="Original")
plt.hist(equalized.ravel(), bins=256, range=(0,256), color='black', alpha=0.6, label="Equalized")

plt.title("Histogram Comparison")
plt.legend()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('orange.jpg')  # sample image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,7))
plt.subplot(2,3,1); plt.imshow(img); plt.title("BGR Image"); plt.axis('off')
plt.subplot(2,3,2); plt.imshow(img_rgb); plt.title("Original (RGB)"); plt.axis('off')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(2,3,3); plt.imshow(gray, cmap='gray'); plt.title("Grayscale"); plt.axis('off')

edges = cv2.Canny(gray, 100, 200)
plt.subplot(2,3,4); plt.imshow(edges, cmap='gray'); plt.title("Canny Edges"); plt.axis('off')

plt.show()
```

```python
from skimage import data, filters
import matplotlib.pyplot as plt   # <-- missing line

img = data.coins()
edges = filters.sobel(img)
blurred = filters.gaussian(img, sigma=2)

fig, ax = plt.subplots(1, 3, figsize=(12,4))
ax[0].imshow(img, cmap='gray'); ax[0].set_title("Original")
ax[1].imshow(edges, cmap='gray'); ax[1].set_title("Sobel Edges")
ax[2].imshow(blurred, cmap='gray'); ax[2].set_title("Gaussian Blur")

for a in ax:
    a.axis('off')

plt.show()
```
