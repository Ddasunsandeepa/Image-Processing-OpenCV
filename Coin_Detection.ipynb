```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img.jpg',0)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()

print(img[100,100])
```

```python
ret,thresh = cv2.threshold(img,200,225,cv2.THRESH_BINARY)
plt.imshow(thresh,cmap='gray')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('img.jpg')
imGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(imGray,cmap='gray')
plt.axis('off')
plt.show()
```

```python
ret,imThresh = cv2.threshold(imGray,200,225,cv2.THRESH_BINARY)
plt.imshow(imThresh,cmap='gray')
plt.axis('off')
plt.show()
```

```python
contours,hierachy = cv2.findContours(imThresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
```

```python
cv2.drawContours(img,contours,-1,(0,255,255),2)


fig,ax= plt.subplots(1,2,figsize=(10,5))

ax[0].imshow(imThresh,cmap='gray')
ax[0].set_title("Image 1")
ax[0].axis("off")

ax[1].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax[1].set_title("Image 2")
ax[1].axis("off")

plt.show()
```
